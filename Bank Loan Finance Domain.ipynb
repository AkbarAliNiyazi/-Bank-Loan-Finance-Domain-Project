# Loan status distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x="loan", data=df, palette="coolwarm")
plt.title("Loan Status Distribution")
plt.xlabel("Loan Taken")
plt.ylabel("Count")
plt.show()
```

# Housing Loan vs Personal Loan Comparison

```python
plt.figure(figsize=(10,5))
sns.countplot(x="loan", hue="housing", data=df, palette="viridis")
plt.title("Comparison of Housing Loan vs Personal Loan")
plt.xlabel("Personal Loan Status")
plt.ylabel("Count")
plt.show()
```

# Loan Distribution Across Different Jobs

```python
plt.figure(figsize=(14,6))
sns.countplot(x="job", hue="loan", data=df, palette="coolwarm", order=df["job"].value_counts().index)
plt.xticks(rotation=90)
plt.title("Loan Status Across Different Job Types")
plt.xlabel("Job Type")
plt.ylabel("Count")
plt.show()
```

# Balance Analysis for People Who Took Loans

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="loan", y="balance", data=df, palette="pastel")
plt.ylim(-5000, 20000)
plt.title("Balance Distribution for People with and without Loans")
plt.xlabel("Personal Loan Status")
plt.ylabel("Balance Amount")
plt.show()
```

# Loan Approval Rate Based on Marital Status

```python
plt.figure(figsize=(10,5))
sns.countplot(x="marital", hue="loan", data=df, palette="Set2")
plt.title("Loan Approval Rate Based on Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.show()
```

# Loan vs Education Level

```python
plt.figure(figsize=(10,5))
sns.countplot(x="education", hue="loan", data=df, palette="coolwarm")
plt.title("Loan Distribution Based on Education Level")
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.show()
```

# Age vs Loan Distribution

```python
plt.figure(figsize=(12,5))
sns.histplot(data=df, x="age", hue="loan", multiple="stack", bins=30, palette="viridis")
plt.title("Age vs Loan Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```
